# LIBRARY MANAGEMENT SYSTEM 

```python
print("WELCOME TO THE LIBRARY MANAGEMENT SYSTEM\n\n")

library={"python_basics":{"author":"james","available":True},
         "Data Science":{"author":"peter","available":True},
         "Cyber Security":{"author":"steve","available":False},
         "Robotics":{"author":"Tony","available":False}
    
}



def lib():
    count=0
    for books,author_info in library.items():
        if author_info["available"]==True :
            
            status="Available"
            count+=1
    
        else:
            status="Not available"
            count+=1
    
        print(f"{count}.Book name    :{books}")
        print(f"  Author name  :{author_info['author']}")
        print(f"  Availability :{status}\n")


# add books
def add():
    while True:
        book=input("enter book name for adding").strip()

        if book in library:
            print("book already exist")

        else:
            author=input("enter author name").strip()
            availability=input("is the book available? enter yes/no").strip().lower()
            if availability=="yes":
                status=True

            else:
                status=False

            library[book]={
                "author":author,"available":status
            }
            print(f"new book {book} added successfully")

        more=input("do u want to add more books?Yes/No").strip().lower()
        if more!="yes":
            break
            

    print("Books available in the library:\n")



# Issue book:
issued_book={}
def issue_book():
    book=input("enter the book which you want to issue")
    name=input("enter the name of the person to whom book is to be issued")

    if book in library:
        if library[book]["available"]:
            library[book]["available"]=False
            issued_book[book]=name
            print(f"Book name:{book} issued by {name}")
            

        else:
            print(f"Book name:{book} already issued")

    else:
        print("book not found")
       




# Return book

def return_book():
    book=input("enter the book name which is to be returned")
    name2=input("enter the name of the person who returned the book")
    if book in library:
        if library[book]["available"]==False:
            library[book]["available"]=True
            if book in issued_book:
                del issued_book[book]
            print(f"Book name:{book} returned successfully by {name2}")

        else:
            print("book is available")

    else:
        print("Book name not found")


# View books

def view_issued_books():
    
    if issued_book=={}:
        print("No books are issued")

    else:
        count=1
        for book in issued_book:
            name=issued_book[book]
            print(f"{count}.Book name:{book} issued by {name}")
            count+=1
            
# Main Menu:

def menu():
    while True:
        choice=input("select an option:\n1.View Books\n2.Add Book\n3.Issue Book\n4.Return Book\n5.View issued books\n6.Exit").strip()
        if choice=="1":
            lib()
    
        elif choice=="2":
            add()
    
        elif choice=="3":
            issue_book()
    
        elif choice=="4":
            return_book()
    
        elif choice=="5":
            view_issued_books()
    
        elif choice=="6":
            print("Thanks for using LMS platform")
            print("Exiting LMS")
            break
    
        else:
            print("invalid choice")
            
            
# login credentials:

credentials={"admin":"admin123",
            "librarian":"lib123"}
def login():
    print("Login in to LMS platform")
        
    attempts=3
    while attempts>0:
        user=input("enter user name")
        password=input("enter password")
    
        if user in credentials and credentials[user]==password:
            print(f"{user} logged in successfully")
            return True
    
        else:
            attempts-=1
            print(f"wrong password!,you have {attempts} attempts left")
    
    print("you enter too many wrong attempts ! pls try again later")
    return False



if login():
    menu()
            
    
                
       
        
```
